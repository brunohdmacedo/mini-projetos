```python
from geopy.geocoders import Nominatim
```

```python
# Parque Nacional da Tijuca - Alto da Boa Vista, Rio de Janeiro - RJ
locator = Nominatim(user_agent="myGeocoder")
location = locator.geocode("Parque Nacional da Tijuca - Alto da Boa Vista, Rio de Janeiro - RJ")
```

```python
import pandas as pd
```

```python
vacina_AM =pd.read_csv('/content/drive/MyDrive/Colab Datasets/vacina-covid/part-00000-5bff8028-67f4-4b92-9914-e0d31a0adf3c.c000.csv',sep=';')
```

```python
vacina_AM.head()
```

```python
vacina_AM_paciente = vacina_AM[['paciente_endereco_nmMunicipio','paciente_endereco_uf','paciente_endereco_nmPais']]
```

```python
vacina_AM_paciente.drop_duplicates(inplace=True)
```

```python
vacina_AM_paciente
```

```python
vacina_AM_paciente['paciente_endereco'] = vacina_AM_paciente['paciente_endereco_nmMunicipio'] + ',' + \
                vacina_AM_paciente['paciente_endereco_uf'] + ',' + \
                vacina_AM_paciente['paciente_endereco_nmPais']
```

```python
vacina_AM_paciente
```

```python
from geopy.extra.rate_limiter import RateLimiter
```

```python
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
```

```python
vacina_AM_paciente['paciente_endereco_geolocalizado'] = vacina_AM_paciente['paciente_endereco'].apply(geocode)
```

```python
vacina_AM_paciente['paciente_endereco_geolocalizado']
```

```python
vacina_AM_paciente['paciente_endereco_latitude'] = vacina_AM_paciente['paciente_endereco_geolocalizado'].apply(lambda loc: loc.latitude if loc else None)
vacina_AM_paciente['paciente_endereco_longitude'] = vacina_AM_paciente['paciente_endereco_geolocalizado'].apply(lambda loc: loc.longitude if loc else None)
```

```python
vacina_AM_paciente['paciente_endereco_point']=vacina_AM_paciente['paciente_endereco_geolocalizado'].apply(lambda loc: tuple(loc.point) if loc else None)
```
