```
us_elections = pd.read_csv("/content/drive/My Drive/Colab Datasets/eleicoes/1976-2016-president.csv")
```

```
selected_year = 2008
us_elections_sel_year = us_elections[us_elections.year == selected_year]
```

```
us_elections_sel_year.drop_duplicates(subset=['state'])
```

```
us_elections_sel_year.sort_values(by=['state','candidatevotes']).drop_duplicates(subset=['state'],keep='last')
```

```
df = us_elections_sel_year.sort_values(by=['state','candidatevotes']).drop_duplicates(subset=['state'],keep='last')
```

```
df['party'] = df['party'].astype('category')
```

```
df.info()
```

```
# df['candidate'].cat.codes
```

```
import plotly.graph_objects as go

# df = us_elections_sel_year.drop_duplicates(subset=['state'])

fig = go.Figure(data=go.Choropleth(
    locations=df['state_po'], # Spatial coordinates
    z = df['party'].cat.codes, # Data to be color-coded
    locationmode = 'USA-states', # set of locations match entries in `locations`
    colorscale = [[0.0, "rgb(0,0,184)"],[1.0, "rgb(184,0,0)"]],
    colorbar_title = "Winner Candidate",
    text = df['candidate']
))

fig.update_layout(
    title_text = '{} US President Election Winner by State'.format(selected_year),
    geo_scope='usa', # limite map scope to USA
)

fig.show()
```
